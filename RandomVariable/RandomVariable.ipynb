### Домашнее задание к лекции.

Смоделировать игру против лотерейного автомата типа “777”. Игрок платит 1 руб., после чего выпадает случайное целое   число, равномерно распределенное от 0 до 999. При некоторых значениях числа игрок получает выигрыш (см. справа)  

Выгодна ли игра игроку?  

Сколько в среднем приобретает  
или теряет игрок за одну игру?  

777: 200 руб.  
999: 100 руб.  
555: 50 руб.  
333: 15 руб.  
111: 10 руб.  
-77: 5 руб.  
--7: 3 руб.  
-00: 2 руб.  
--0: 1 руб.  

/   * – любая цифра  

```python
import numpy as np
import random
```

```python
# рассмотрим два типа игроков:

# первый тип: 
# у игрока есть определенная сумма, к примеру 1000 руб
# мы знаем, что один кон игры стоит 1 руб
# если игрок выигрывает - он откладывает эти деньги и не тратит на новую игру
# то есть, сумма, которая была у него в самом начале игры с каждым коном уменьшается на 1 руб
# при этом, у игрока заканчивается начальная сумма, он перестает играть

def game_777_1(free_money):
    win_money = 0 #сколько денег выиграл
    win_game = 0 #сколько игр выиграл

    lost_money = 0 #сколько денег потеряно (начальная сумма минус 1 рубль)
    lost_game = 0 #проигранные игры
    
    total_games = 0 #всего игр
    # free_money сколько изначально было денег у игрока
    
    win_number = {'777': 200, '999': 100, '555': 50, '333': 15, '111': 10} #выиграшные числа
    win_remainder = {'77': 5, '7': 3, '00': 2, '0': 1} #выиграшные остатки от чисел
    
    for i in range(free_money):
        lost_money += 1 #деньги расходуются при каждом выигрыше, так же, как и проигрыше
        random_num = str(int(round(random.uniform(0, 1000), 0))) #случайное число по условию задачи от 0 до 999 (всего 1000 чисел)
        if random_num in win_number.keys():
            win_money += win_number[random_num]
            win_game += 1
        elif random_num[1:] in win_remainder.keys():
            win_money += win_remainder[random_num[1:]]
            win_game += 1
        elif random_num[2:] in win_remainder.keys():
            win_money += win_remainder[random_num[2:]]
            win_game += 1
        else:
            lost_game += 1
    else:
        lost_game += 1
            
    print('Сумма выигранных денег: '  +  str(win_money) + '\n'
          'Сумма игр с выигрышем: ' +  str(win_game)  + '\n'
          'Истрачено денег на игру: '  +  str(lost_money) + '\n'
          'Проигранных игр: '  +  str(lost_game) + '\n'
#  сравним теоретическую вероятность до эксперимента и частотой, которую мы получим после проведения эксперимента
          'Теоретическая вероятность: ' + str((int(str(1000)[:2]) + int(str(1000-100)[:2]) + int(str(1000)[:2])+ int(str(1000-100)[:2]))/1000) + '\n'
          'Метод Монте-Карло: ' + str(win_game/(lost_game + win_game)) + ' (выигранные игры / все игры)')
#     return win_money

# посчитаем теоретическую вероятность данной игры
# так как вероятность одного хода игры не зависит от количества экспериментов (в данном случае денег),
# (каждая игра отнимает 1 рубль),
# а зависит от того, из какого количества чисел будет выбираться случайное число
# посчитаем количество выигрышных чисел 
# '777': 200, '999': 100, '555': 50, '333': 15, '111': 10 = 5 возможных чисел 
# '77': 5, '7': 3, '00': 2, '0': 1 = 198 возможных чисел
# int(str(1000)[:2]) #*77 = 9
# int(str(1000-100)[:2]) #*7 = 90
# int(str(1000)[:2]) #*00 = 9
# int(str(1000-100)[:2]) #**0 = 90
```

```python
game_777_1(1000000)
#видим, что теоретическая вероятность близка к частоте, которую мы получили в ходе эксперимента 
```

```python
# Сколько в среднем приобретает или теряет игрок за одну игру?

# Среднее кол-во денег, которое можно выиграть за игру:
# Посчитаем вероятность выпадения каждого из возможных чисел и умножим на сумму денег, которая есть в начале игры
# и далее умножим на сумму, которую можно получить при выподении данного числа
```

```python
# Для чисел '777': 200, '999': 100, '555': 50, '333': 15, '111': 10   представим их, как вероятность 1/1000, 
# так как может быть только одно такое число за один кон игры

# для чисел '77': 5, '7': 3, '00': 2, '0': 1 представим их, как вероятность всех возможных появлений таких чисел
# int(str(1000)[:2]) #*77 = 9
# int(str(1000-100)[:2]) #*7 = 90
# int(str(1000)[:2]) #*00 = 9
# int(str(1000-100)[:2]) #**0 = 90
```

```python
def win_money(sum_money):
    
    win_number_sum = (1/1000*sum_money*200) + \
                     (1/1000*sum_money*100) + \
                     (1/1000*sum_money*50) + \
                     (1/1000*sum_money*15) + \
                     (1/1000*sum_money*10)
    
    win_remainder_sum = (int(str(1000)[:2])/1000*sum_money*5) + \
                        (int(str(1000-100)[:2])/1000*sum_money*3) + \
                        (int(str(1000)[:2])/1000*sum_money*2) + \
                        (int(str(1000-100)[:2])/1000*sum_money*1)
    return win_number_sum + win_remainder_sum
win_money(100000)
```

```python
# второй тип игроков: 
# у игрока есть определенная сумма, к примеру 1000 руб
# мы знаем, что один кон игры стоит 1 руб
# если игрок выигрывает - он тратить выирыш на новую игру
# то есть, сумма, которая была у него в самом начале игры с каждым коном уменьшается на 1 руб и увеличивается на 
# сумму выигрыша
# Играет до тех пор, пока есть деньги на игру


def game_777_2(free_money):
    win_money = 0 # сколько денег выиграл
    win_game = 0 # сколько игр выиграл

    lost_money = 0 # сколько денег потеряно (начальная сумма минус 1 рубль)
    lost_game = 0 # проигранные игры
    
    total_games = 0 # всего игр
    # free_money сколько изначально было денег у игрока
    
    win_number = {'777': 200, '999': 100, '555': 50, '333': 15, '111': 10} #выиграшные числа
    win_remainder = {'77': 5, '7': 3, '00': 2, '0': 1} #выиграшные остатки от чисел
    
    while free_money > 0:
        random_num = str(int(round(random.uniform(0, 999), 0))) #случайное число по условию задачи от 0 до 999 (всего 1000 чисел)
        lost_money += 1 #деньги расходуются при каждом выигрыше, так же, как и проигрыше
        if random_num in win_number.keys():
            win_money += win_number[random_num]
            win_game += 1
            free_money += win_number[random_num]
        elif random_num[1:] in win_remainder.keys():
            win_money += win_remainder[random_num[1:]]
            win_game += 1
            free_money += win_remainder[random_num[1:]]
        elif random_num[2:] in win_remainder.keys():
            win_money += win_remainder[random_num[2:]]
            win_game += 1
            free_money += win_remainder[random_num[2:]]
        else:
            lost_game += 1
            free_money -= 1
    else:
        lost_game += 1
        free_money -= 1
            
    print('Сумма выигранных денег: '  +  str(win_money) + '\n'
          'Сумма игр с выигрышем: ' +  str(win_game)  + '\n'
          'Истрачено денег на игру: '  +  str(lost_money) + '\n'
          'Проигранных игр: '  +  str(lost_game) + '\n'
          'Теоретическая вероятность: ' + str((int(str(1000)[:2]) + int(str(1000-100)[:2]) + int(str(1000)[:2]) + int(str(1000-100)[:2]))/1000) + '\n'
          'Метод Монте-Карло: ' + str(win_game/(lost_game + win_game)) + ' (выигранные игры / все игры)')
    #return win_money
```
