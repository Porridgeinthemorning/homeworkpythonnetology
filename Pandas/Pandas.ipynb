### Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4,  12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
file = pd.read_csv('power.csv')
```

```python
file[file['country'].str.contains(('nia'), case=False)]['country'].unique()
```

```python
df = file[((file['country'] == 'Latvia') 
         | (file['country'] == 'Lithuania') 
         | (file['country'] == 'Estonia') 
         | (file['country'] == 'Latvia'))
         & (file['quantity'] >= 0)
        & ((file['category'] == 4)
         | (file['category'] == 12)
         | (file['category'] == 21))
        & ((file['year'] >= 2005)
         & (file['year'] <= 2010))]
```

```python
df = df.groupby(['country', 'year', 'category']).sum()
df = df.sort_values(by=['country', 'quantity'], ascending=False)
```
